```python
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]

profit_a=[(a[i]-a[i-1]) for i in range(i,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(i,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',market='>',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='CompanyB',market='H')
plt.show()

```
