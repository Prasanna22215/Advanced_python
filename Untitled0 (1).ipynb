```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a=pd.read_csv("/content/Temperature.txt",sep=" ")
print(a)
print("\n the average in 50 days is:")
avg=a['Temperature'].mean()
print(a)
print("\nthe lowest temperature recorded is:")
print(a['Temperature'].min())
print("\nthe highest temperature recorded is:")
print(a['Temperature'].max())
day=0
for i in a['Temperature']:
  if(i>avg):
    day=day+1
    print("\nnumber of days where the temperature exceeded threshold is:\n")
    plt.plot(a['Days'],a['Temperature'],color='black')
    plt.xlabel('Days')
    plt.ylabel('Temperature')
    plt.show()
```
